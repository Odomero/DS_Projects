```python
import os

def get_filepaths(directory):
    """
    This function will generate the file names in a directory 
    tree by walking the tree either top-down or bottom-up. For each 
    directory in the tree rooted at directory top (including top itself), 
    it yields a 3-tuple (dirpath, dirnames, filenames).
    """
    file_paths = []  # List which will store all of the full filepaths.

    # Walk the tree.
    for root, directories, files in os.walk(directory):
        for filename in files:
            # Join the two strings in order to form the full filepath.
            filepath = os.path.join(filename)
            file_paths.append(filepath)  # Add it to the list.

    return file_paths  # Self-explanatory.

# Run the above function and store its results in a variable.   
file_names = get_filepaths("/Users/Mrs Omokhafe/Documents/data")
```

```python
from tqdm import tqdm
import pandas as pd
```

```python
"""
This loop reads the content of different files
into a DataFrame and apend to a single list.
"""
  
frame = []
for f in tqdm(file_names):
    path = 'data/' + f
    var = f.split('_')[0]
    col_name = var.split('#')
    col_name = pd.read_csv(path)
    frame.append(col_name)
print(frame)
```

```python
#merging all DataFrames in the list(frame)

result = pd.concat(frame)
```

```python
"""
This loop reads the content of different files
into a DataFrame and apend to a dictionary with 
file-name as the key and DataFrame as the value.
"""
frame = {}
frames = []
name_list = []
for f in tqdm(file_names):
    path = 'data/' + f
    var = f.split('_')[0]
    name = var.split('#')[1]
    name_list.append(name)
    df = pd.read_csv(path)
    frames.append(df)
    
for fname, file in zip(name_list, frames):
        frame[fname] = frame.get(fname, file)
print(frame)
```

```python
#merging all DataFrames in the dictionary(frame)

final_df = pd.concat(frame)
```

```python
def getHashtags(tweet):
    """
    This function extracts hashtags
    from tweets
    """
    tweet = tweet.lower()  #has to be in place
    tweet = re.findall(r'\#\w+',tweet) # Remove hastags with REGEX
    return " ".join(tweet)
```
